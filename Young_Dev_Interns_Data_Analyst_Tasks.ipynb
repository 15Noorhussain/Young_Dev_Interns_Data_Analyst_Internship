###Task 2 DATA Visualization with Matplotlib/Seaborn


1.   Create a Histogram to visualize the distribution of a numerical variable.
2.   Plot a bar chart to show the frequency of categories in a categorical variable.
3.   Create a scatter plot to visualize the relationship between two numerical variables.







```python
###Task 2 Part 1 Histogram
numerical_variable = df['SALARY']
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(numerical_variable, kde=True, color='skyblue')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Histogram of Salary Distribution')
plt.show()
```

###Plotting a Bar chart

```python
categorical_variable = df['DESIGNATION']
category_counts = categorical_variable.value_counts()
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Bar Chart of Categorical Variable')
plt.show()
```

###Scatter Plot for two numerical variables

```python
variable1 = df['SALARY']
variable2 = df['PAST EXP']
plt.figure(figsize=(10, 6))
plt.scatter(variable1, variable2, color='blue', alpha=0.7)
plt.figure(figsize=(10, 6))
plt.scatter(variable1, variable2, color='blue', alpha=0.7)  # alpha controls transparency

# Add labels and title
plt.xlabel('SALARY')
plt.ylabel('PAST EXP')
plt.title('Scatter Plot of Relationship between SALARY and PAST EXPERIENCE')

# Show the plot
plt.show()
```

###Task 4 Time Series Analysis


1.   Analyze time series data, including trend and seasonality decomposition.
2.   Forecast future values using time series forecasting methods



```python
from statsmodels.tsa.seasonal import seasonal_decompose #Import necessary libraries
```

```python
time_series = pd.read_csv('/content/drive/MyDrive/Time Series Dataset/daily-minimum-temperatures-in-me.csv') #Loading the Dataset
```

```python
time_series.head() #Examining the new Time series Dataset
```

```python
time_series.columns #Checking different columns
```

```python
time_series.rename(columns={'Daily minimum temperatures': 'Temp'}, inplace=True) #Renaming the Column name
```

```python
time_series['Date'] = pd.to_datetime(time_series['Date']) #Converting into datetime format
time_series.set_index('Date', inplace=True)
```

```python
time_series.head() #Rechecking the Dataset
```

```python
non_numeric_values = time_series['Temp'].apply(lambda x: not x.isnumeric())
non_numeric_rows = time_series[non_numeric_values]
```

```python
print(time_series.shape)
print(non_numeric_rows.shape)
```

```python
print(time_series.index)
print(non_numeric_rows.index)
```

```python
time_series.drop(non_numeric_rows.index, inplace=True)
```

```python
import matplotlib.pyplot as plt
result = seasonal_decompose(time_series['Temp'], model='additive', period=12) #Plotting the Orignal time series
plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.plot(time_series['Temp'], label='Original Time Series')
plt.legend()
```

```python
plt.subplot(4, 1, 2) #Plotting the Trend Component
plt.plot(result.trend, label='Trend Component')
plt.legend()
```

```python
plt.subplot(4, 1, 3) #Plotting the Seasonal Component
plt.plot(result.seasonal, label='Seasonal Component')
plt.legend()
```

```python
plt.subplot(4, 1, 4) #Checking the Residuals
plt.plot(result.resid, label='Residuals')
plt.legend()
plt.tight_layout()
plt.show()
```
